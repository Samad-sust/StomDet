````python

# Import the modules 

from ultralytics import YOLO
````

````python
# Train the model 
# yolo task=detect mode=train epochs=20 data=data_custom.yaml model=yolov8m.pt imgsz=640

# model = YOLO('./yolov8s.pt')
model = YOLO('yolov8n.yaml')

data = './dataset.yaml'
````

````python
# model.train(data=data, batch=8, epochs=150, workers=4, imgsz=608, 
#             patience=10, single_cls=True, lr0=0.0053, weight_decay=0.0078, 
#             dropout=0.013, retina_masks=False, augment=False)

model.train(data=data, batch=8, epochs=150, workers=4, imgsz=608, 
            patience=10, single_cls=True,)
````

````python
# Predict the images or videos

# yolo task=detect mode=predict model=yolov8_custom.pt show=True 
#   conf=0.5 source=1.jpg save_txt=True save=True

trained_model = YOLO("./runs/detect/train/weights/best.pt")

test_image = "./Images/test"
trained_model.predict(conf=0.5, source= test_image, save=True, save_txt=True, show_labels=False, line_width=3)
````
